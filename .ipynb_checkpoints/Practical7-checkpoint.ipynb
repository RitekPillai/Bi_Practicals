```python
import pandas as pd
```

```python
df = pd.read_csv("SampleSuperstore.csv")
```

```python
print("First 5 rows of the dataset" ,df.head())
```

```python
print("Data set information")
print(df.info())#shows  column names and missing value
```

```python
print('statistical summary')
df.describe()#do the mean,meidan mode and avg of the data set
```

```python
print("Total Sales ",df['Sales'].sum())
print("Total Quantity ",df['Quantity'].sum())
print("Total Profit ",df['Profit'].sum())
```

```python
category_sales=df.groupby('Category')['Profit'].sum() #
```

```python
print("Total profit of each category ",category_sales)
```

```python
top_customers = df.sort_values(by='Sales', ascending=False).head(5)
print(top_customers)
```
